## Costos marginales

```r
#Datos de los costos marginales reales observados
df_cm = read.csv("costo_marginal_real.csv")

#Datos de los costos marginales previstos por el sistema 
df_cp = read.csv("costo_marginal_programado.csv")
```

```r
costo_marginal = merge(df_cm, df_cp, by.x = c('barra_mnemotecnico', 'fecha', 'hora'),by.y = c('mnemotecnico_barra', 'fecha', 'hora'))
```
